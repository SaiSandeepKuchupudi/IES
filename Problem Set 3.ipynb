```python
#step1:
#Importing required libraries

import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
```

```python
#step2 :
#Importing dataset

users = pd.read_csv("C:\\Users\\kuchi\\employee.csv")
users.head()
```

```python
#step4 :
#Mean age by oocupation

users.groupby(['occupation'])['age'].mean()
```

```python
#step5 :
#Male ratio per occupation

crosstab=pd.crosstab(users.occupation, users.gender,normalize='index')
crosstab.drop(["F"], axis = 1, inplace = True) 
crosstab.sort_values(by='M', ascending=False)
```

```python
#step 6:
#Minimum age by oocupation

users.groupby(['occupation'])['age'].min()
```

```python
#step 6:
#Maximum age by oocupation

users.groupby(['occupation'])['age'].max()
```

```python
#step 7:
#Mean age by oocupation and sex

round(users.groupby(['occupation', 'gender'])['age'].mean(), 2)
```

```python
#step 8:

#Male and female percentage per occupation

crosstab=round(pd.crosstab(users.occupation, users.gender,normalize='index')*100, 2)
crosstab.sort_values(by='M', ascending=False)
```
